```python
RGGI_Status = []

for x in states:
    if x == 'Connecticut' or x == 'Delaware' or x == 'Maine' or x == 'Maryland' or x == 'Massachusetts' or x == 'New Hampshire' or x == 'New Jersey' or x == 'New York' or x == 'Pennsylvania' or x == 'Rhode Island' or x == 'Vermont':
        RGGI_Status.append(1)
    else:
        RGGI_Status.append(0)

data['RGGI'] = RGGI_Status
```

```python
unemployment = pd.pivot_table(data, values = 'Value', index = 'State', columns = 'Year', aggfunc = 'mean')
unemployment
```

```python
unemployment_index = unemployment.reset_index()
```

```python
unemployment_index
```

```python
# Extract the row you want to plot
row_to_plot = unemployment.loc['Wisconsin']

# Plot the row as a line graph
row_to_plot.plot(kind='line', marker='o')
#plt.title('Line Graph of Category A and B for Year 2020')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
rggi_statuses = pd.pivot_table(data, values = 'Value', index = 'RGGI', columns = 'Year', aggfunc = 'mean')
rggi_statuses
```

```python
rggi_statuses.T.plot(figsize=(12,6))
plt.title('Unemployment Rate for RGGI States vs. Non-RGGI States')
plt.xlabel('Year')
plt.ylabel('Seasonal Adjusted Unemployment Rate')
plt.xticks(range(2006,2025))
plt.axvline(x=2009, color='red', linestyle='--', linewidth=2)
#plt.grid(True)
plt.legend(['Non-RGGI', 'RGGI'])
plt.show()
```
